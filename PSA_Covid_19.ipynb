```python
# Doing our Analysis 
ncov_record = SeqIO.read("sequence.fasta","fasta")
```

```python
ncov_record
```

```python
ncov_dna = ncov_record.seq
```

```python
# covid DNA
ncov_dna
```

```python
# Len of our sequence
len(ncov_dna)
```

```python
# protein synthesis
# DNA -> mRNA -> Protein/AA
```

```python
# Transcription (DNA to mRNA)
ncov_mrna = ncov_dna.transcribe()
```

```python
ncov_mrna
```

```python
# Translate to Amino Acids/ Protein 
ncov_protein = ncov_mrna.translate()
```

```python
ncov_protein
```

```python
len(ncov_protein)
```

```python
# Custom 
# Condon = 3
len(ncov_dna)/3
```

```python
# find all AA sequence before stopcondon
ncov_protein
```

```python
ncov_aa = ncov_protein.split("*")
```

```python
ncov_aa
```

```python
ncov_clean = [str(i) for i in ncov_aa]
```

```python
ncov_clean
```

```python
# Using Pandas
import pandas as pd
```

```python
df = pd.DataFrame({'amino_acids':ncov_clean})
```

```python
df['count'] = df['amino_acids'].str.len()
```
